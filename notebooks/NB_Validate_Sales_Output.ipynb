```python
# Validation post-exécution des données analytiques

df = spark.table("sales_orders_clean")

# Vérifier que la table contient des données
row_count = df.count()
if row_count == 0:
    raise Exception("Validation failed: sales_orders_clean is empty")

# Vérifier la cohérence temporelle
max_date = df.selectExpr("max(order_date) as max_date").collect()[0]["max_date"]
if max_date is None:
    raise Exception("Validation failed: no valid order_date found")

print("Validation OK")
print("Rows:", row_count)
print("Max order_date:", max_date)
```
